# Problem 2

We want the steady-state temperature dsitribution in a rectangular copper plate. This distribution satisfies:

$$\frac{\partial^2 u}{\partial x^2} + \frac{\partial^2 u}{\partial y^2} = 0$$

Where on the upper and lower boundaries, the second term $\frac{\partial^2 u}{\partial y^2}=0$, the left side is at 0C, and the right is at $f(y)=y$C. 

The analytical solution computed using separation of variables is:

$$u(x,y)=\frac{x}{4} - 4\sum_{n=1}^\infty \frac{\sinh(n\pi x)\cos(n\pi y)}{(n\pi)^2\sinh(2n\pi)}$$
Where $n$ is odd ($n' = 2n+1$, from $n=0$ to $n=\infty$). 

We want to write a program that computes the steady state using Jacobi, Gauss-Seidel, and SOR, with $N_x$ cells ($N_x+1$ pts) in the x direction, and $N_y$ cells ($N_y+1$ pts) in the y direction. We use a first order approximation for the Neumann boundaries. My interpretation of this is to use the following FDA's for the bottom and top, respectively:

$$\frac{\partial u}{\partial y} = \frac{u(1)-u(0)}{\Delta y}$$
$$\frac{\partial u}{\partial y} = \frac{u(N_y+1)-u(N_y)}{\Delta y}$$

We initialize the interior with temperature 0C.

To determine convergence, we define a residual:
$$res = \sqrt{\left(\sum_{i=2}^{N_x-1}\sum_{j=2}^{N_y-1}\left[u^{(k+1)}(i,j)-u^{(k)}(i,j)\right]^2\right)/\left((N_x-2)(N_y-2)\right)}$$

With a tolerance of $10^{-6}$.

We use a contour plot to compare the numerical and analytical isotherms.

```python
# Step 1: Define the problem

#21 x 21 mesh
Nx = 21
Ny = 21

#length 2, height 1
L=2
H=1

#num points
numptsx = Nx+1
numptsy = Ny+1

#dx, dy
dx = L/Nx
dy = H/Ny

# u vector 
# (should be a single column, so that I can multiply the solution by a matrix? Do I need to do that?)
# Would be nice to keep it as a 2D array...
# Ultimately, I decide to keep it as a 2D array
```

```python
# Step 2: define iteration methods
# take in solution vector u^k, dx, and dy
# output the new vector u^k+1

# Jacobi
def jacobi(umatr,omega): 
    uk1 = np.full_like(umatr,0)
    uk1[:,0] = umatr[:,0]
    uk1[:,numptsx-1] = umatr[:,numptsx-1]
    for i in range(1,numptsx-1): #0 and numptsx-1, the endpoints, are defined and shoudl be excluded
            for j in range(1,numptsy-1): # 0 and numptsy-1, the endpoints, should be filled by von neumann BC's
                uk1[i,j] = ( (umatr[i-1,j] + umatr[i+1,j])/dx**2 + (umatr[i,j-1] + umatr[i,j+1])/dy**2 ) * 1/(2/dx**2 + 2/dy**2)
                # if i == numptsx-2:
                #     print(uk1[i,j])
                #     print(umatr[i,j])
    return uk1

# Gauss-Seidel
def gs(umatr,omega):
    uk1 = np.copy(umatr)
    for i in range(1,numptsx-1): #0 and numptsx-1, the endpoints, are defined
            for j in range(1,numptsy-1): # 0 and numptsy-1, the endpoints, should be filled by von neumann BC's
                #at every point that has the new values, use the new values.
                # Let's go from the bottom left. The bottom left interior point will have no new values. 
                # Then, bottom left interior + dx has 1 new value to the left, so we can use that one. 
                #  + dx + dy (3rd,3rd) => (2,2)has one new value on bottom and one to the left
                # In general, there will only ever be new values to the left 
                # New values to the left do not exist at (:,2nd) => (:,1), where the left values are the Dirchlet BC
                # New values on bottom do not exist at (2nd,:) => (1,:) where the bottom values are the von neumann BC's
                uk1[i,j] = ( (uk1[i-1,j] + uk1[i+1,j])/dx**2 + (uk1[i,j-1] + uk1[i,j+1])/dy**2 ) * 1/(2/dx**2 + 2/dy**2)
                # Here, we just overwrite the previous values whenever we calculate a new value.
    return uk1
                
# SOR
def SOR(umatr,omega):
    uk1 = np.copy(umatr)
    for i in range(1,numptsx-1):
        for j in range(1,numptsy-1):
            uk1[i,j] = (1-omega)*umatr[i,j] + omega*( (uk1[i-1,j] + uk1[i+1,j])/dx**2 + (uk1[i,j-1] + uk1[i,j+1])/dy**2 ) * 1/(2/dx**2 + 2/dy**2)
    return uk1
```

```python
# Step 3: fill the von neumann boundary conditions on top and bottom
def vonneumann(uk1):
    # print(uk1)
    # fill top:
    uk1[0,1:numptsx-1] = uk1[1,1:numptsx-1]
    # print(uk1)
    # fill bottom
    uk1[numptsy-1,1:numptsx-1] = uk1[numptsy-2,1:numptsx-1]
    return uk1
```

```python
# Step 4: define convergence criteria
def res(uk1,u):
    sumval = 0
    for i in range(1,numptsx):
        for j in range(1,numptsy):
            sumval += (uk1[i,j]-u[i,j])**2
    res = sumval / ((numptsx-2)*(numptsy-2))
    # print('sumval = ' + str(sumval))
    # check coordinates of 1, numptsx-2 are really the coordinates of the interior
    # check numptsx-3^2 is really the number of points in the interior
    return res
```

```python
# Step 5: solve with some convergence criteria

def solve(method,omega=0,printvals=False,gettstep=False):
    u = np.zeros((numptsy,numptsx)) # origin is at bottom left of the rectangle
    uright = np.linspace(0,H,numptsy)
    # uright = np.flip(uright)
    u[:,numptsx-1] = uright
    # print('u matrix = ')
    # print(u)
    #satisfies all dirchlet and von neumann conditions
    resval = 100 # arbitrary value greater than tolerance to initialize the variable
    tstep = 0
    while resval >= 10**(-6):
        tstep += 1
        # tstep is non-physical, since there is no "time" associated with this steady-state solution method
        uk1 = method(u,omega) #iterate from previous step
        uk1 = vonneumann(uk1) # fill von neumann BC's
        # --> should this happen before or after I solve the rest of the field?
        # because it's a loop, and the initial satisfies the von neumann BC's, filling before and after are equivalent
        # print('last 3 columns uk1 = ' + str(uk1[:,numptsx-3:numptsx]))
        resval = res(uk1,u)
        u = uk1 # uk1 from this step becomes u for next step
    if printvals:
        print('time step = ' + str(tstep))
        print('uk1 shape = ' + str(np.shape(uk1)))
        # print('uk1 matrix = ' + str(uk1))
        print('last 3 columns uk1 = ' + str(uk1[:,numptsx-3:numptsx]))
        # print('2nd to last column uk1 = ' + str(uk1[:,numptsx-1]))
        print('res = ' + str(resval))
    if gettstep == False:
        return uk1
    else:
        return uk1, tstep
```

```python
#Step 6: compare to analytical solution

class matrixinfo:
    def __init__(self,u,x,y):
        self.str = "class containing u 2D array, x and y 1D arrays"
        self.u = u
        self.x = x
        self.y = y

def analytic(x,y):
    dxan = 0.05
    dyan = 0.05
    u = np.zeros([np.size(y),np.size(x)])
    numxvals = np.size(x)
    numyvals = np.size(y)
    for xval in x:
        xindex = int(int(xval/dxan))
        for yval in y:
            yindex = int(int(yval/dyan))
            n = 1
            sumval = 0
            # res = np.ndarray([100]) #choose arbitrary large value to initialize res
            # while res.any() >= 10**(-7):
            while n <= 100:
                prev = 0
                sumval = sumval + np.sinh(n*math.pi*xval) * np.cos(n*math.pi*yval) / ((n*math.pi)**2 * np.sinh(2*n*math.pi))
                # res = sumval-prev
                n += 2
                # prev = sumval
            u[yindex,xindex] = xval/4 - 4 * sumval
            # print('sumval residualf = ' + str(res))
            # print(xval)
            # print(yval)
            # print('4*sumval = ' + str(4*sumval))
            # print('n = ' + str(n-2))
    return u
```

```python
def plotcontours(method=jacobi,omega=1,filename=[]):
    xnum = np.arange(0,L+dx/2,dx)
    ynum = np.arange(0,H+dy/2,dy)
    Xnum, Ynum = np.meshgrid(xnum,ynum)
    ufinal = solve(method,omega,printvals=False)
    
    
    plt.figure()
    cp = plt.contour(Xnum, Ynum, ufinal, 5, colors='black', linestyles='dashed')
    plt.clabel(cp, inline=True, fontsize=10)
    plt.title('Temp Contour: ' + filename)
    plt.xlabel('x')
    plt.ylabel('y')
    
    

    dxan = 0.05
    dyan = 0.05
    x = np.arange(0,L+dxan,dxan)
    y = np.arange(0,H+dyan,dyan)
    X, Y = np.meshgrid(x, y)
    Z = analytic(x,y)
    
    cp = plt.contour(X, Y, Z, 5, colors='grey')
    plt.clabel(cp, inline=True, fontsize=10)
    if len(filename) != 0:
        plt.savefig('HW3_PR3_' + filename + '_contour.png')
    plt.show()
    
    plt.figure()
    fig, ax = plt.subplots(subplot_kw={"projection": "3d"})
    surf = ax.plot_surface(X,Y,Z) #light blue
    surf = ax.plot_surface(Xnum,Ynum,ufinal) #orange
    plt.title('3D Visualization of Temperature Distribution: ' + filename)
    plt.xlabel('x')
    plt.ylabel('y')
    if len(filename) != 0:
        plt.savefig('HW3_PR3_' + filename + '_3D.png')
    plt.show()

    return None
```

```python
plotcontours(jacobi,filename='Jacobi')
plotcontours(gs,filename='Gauss_Seidel')
plotcontours(SOR,omega=1.83,filename='SOR_omega=1.83')
```

# SOR Optimization

```python
tstepvec = []
omegavec = np.arange(0,2,0.001)
for omega in omegavec:
    unpack = solve(SOR,omega,printvals=False,gettstep=True)
    tstep = unpack[1]
    # print(tstep)
    tstepvec.append(tstep)
```

```python
plt.figure(figsize=(10,5))
axes_internal = plt.gca()
axes_internal.set_xlabel('$\omega$')
axes_internal.set_ylabel('# Time Steps To Convergence')
# plt.title(title)
plt.plot(omegavec,tstepvec)
plt.savefig('SOR_Convergence.png')
plt.show()
```
