Problem 1: Plot the real and imaginary parts of $k^* \Delta x$ vs $k \Delta x$, for a domain of $0 \leq k \Delta x \leq\pi$
=====================

```python
# I want to plot the real component, the imaginary component, and the magnitude, on the same plot.
# I should make a real component function and a complex component function for each scheme, and a single magnitude function.
# I should make a plotting function to plot three separate values, and label them appropriately.

def plot_vals(x1, exact, y1=[], y2=[], y3=None,filename=[], logscale = False, xlabel='$k\Delta x$', ylabel='$k^*\Delta x$', title=''):
    plt.figure(figsize=(10,5))
    plt.grid()
    axes_internal = plt.gca()
    if logscale: 
        # axes_internal.set_yscale('log')
        axes_internal.set_xscale('log')
    plt.plot(x1,exact, label = 'Exact Solution, $k=k^*$')
    if xlabel: axes_internal.set_xlabel(xlabel)
    if ylabel: axes_internal.set_ylabel(ylabel)
    if title: plt.title(title)
    if np.any(y1): plt.plot(x1,y1, label = '$k^*_{Re}$')
    if np.any(y2): plt.plot(x1,y2, label = '$k^*_{Im}$')
    if np.any(y3): plt.plot(x1,magcomplex(y1,y2), label = 'Magnitude $|k^*|$')
    plt.axvline(math.pi, color='k', label='$k\Delta x=\pi$')
    plt.legend()
    if len(filename) != 0:
        plt.savefig(filename+'.png')
    plt.show()

def magcomplex(re,im):
    return np.sqrt(re**2+im**2)

def kstarS1dxre(kdx):
    return np.sin(kdx)
    
def kstarS1dxim(kdx):
    return np.cos(kdx)-1
    
def kstarS2dxre(kdx):
    return np.sin(kdx)

def kstarS3dxre(kdx):
    return (8*np.sin(kdx)-np.sin(2*kdx))/6

def kstarS3dxim(kdx):
    return -(3+np.cos(2*kdx)-4*np.cos(kdx))/6

def kstarS4dxre(kdx):
    return (8*np.sin(kdx)-np.sin(2*kdx))/6
```

```python
kdx = np.arange(0,math.pi,0.01)
plot_vals(kdx,kdx,kstarS1dxre(kdx),kstarS1dxim(kdx),1,'P1b_S1')
plot_vals(kdx,kdx,kstarS2dxre(kdx),[],None,'P1b_S2')
plot_vals(kdx,kdx,kstarS3dxre(kdx),kstarS3dxim(kdx),1,'P1b_S3')
plot_vals(kdx,kdx,kstarS4dxre(kdx),[],None,'P1b_S4')
```
