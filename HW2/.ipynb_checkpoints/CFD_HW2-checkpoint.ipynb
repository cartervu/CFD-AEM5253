Problem 1: Plot the real and imaginary parts of $k^* \Delta x$ vs $k \Delta x$, for a domain of $0 \leq k \Delta x \leq\pi$
=====================

```python
# I want to plot the real component, the imaginary component, and the magnitude, on the same plot.
# I should make a real component function and a complex component function for each scheme, and a single magnitude function.
# I should make a plotting function to plot three separate values, and label them appropriately.

def plot_vals(x1, exact, y1=[], y2=[], y3=None,filename=[], logscale = False, xlabel='$k\Delta x$', ylabel='$k^*\Delta x$', title=''):
    plt.figure(figsize=(10,5))
    plt.grid()
    axes_internal = plt.gca()
    if logscale: 
        # axes_internal.set_yscale('log')
        axes_internal.set_xscale('log')
    plt.plot(x1,exact, label = 'Exact Solution, $k=k^*$')
    if xlabel: axes_internal.set_xlabel(xlabel)
    if ylabel: axes_internal.set_ylabel(ylabel)
    if title: plt.title(title)
    if np.any(y1): plt.plot(x1,y1, label = '$k^*_{Re}$')
    if np.any(y2): plt.plot(x1,y2, label = '$k^*_{Im}$')
    if np.any(y3): plt.plot(x1,magcomplex(y1,y2), label = 'Magnitude $|k^*|$')
    plt.axvline(math.pi, color='k', label='$k\Delta x=\pi$')
    plt.legend()
    if len(filename) != 0:
        plt.savefig(filename+'.png')
    plt.show()

def magcomplex(re,im):
    return np.sqrt(re**2+im**2)

def kstarS1dxre(kdx):
    return np.sin(kdx)
    
def kstarS1dxim(kdx):
    return np.cos(kdx)-1
    
def kstarS2dxre(kdx):
    return np.sin(kdx)

def kstarS3dxre(kdx):
    return (8*np.sin(kdx)-np.sin(2*kdx))/6

def kstarS3dxim(kdx):
    return -(3+np.cos(2*kdx)-4*np.cos(kdx))/6

def kstarS4dxre(kdx):
    return (8*np.sin(kdx)-np.sin(2*kdx))/6
```

```python
kdx = np.arange(0,math.pi,0.01)
plot_vals(kdx,kdx,kstarS1dxre(kdx),kstarS1dxim(kdx),1,'P1b_S1')
plot_vals(kdx,kdx,kstarS2dxre(kdx),[],None,'P1b_S2')
plot_vals(kdx,kdx,kstarS3dxre(kdx),kstarS3dxim(kdx),1,'P1b_S3')
plot_vals(kdx,kdx,kstarS4dxre(kdx),[],None,'P1b_S4')
```

Problem 3: Develop a program to solve the 1D advection equation:
===
$$\frac{\partial f}{\partial t} + a\frac{\partial f}{\partial x} = 0$$
given an initial condition $f_0(x)$ on a periodic domain with $0\leq x<L$ with $N$ uniformly distributed points. Make your program flexible enough to accommodate switching between different spatial discretizations and different explicit time stepping methods (including at LEAST explicit Euler and RK3). You should be able to change a, L, N, $\Delta t$, and the initial condition $f_0(x)$. 

```python
# First, I want to determine \Delta x, \Delta t, the list of xvals, the list of tvals, and the times $T_p$ that I want to save the solution at.
# Next, I want to represent the spatial scheme.
# I also want to create a matrix $F$ that represents the solution, with the solution at each time step being a column of $F$, F(n,:)
# The size of $f$ should be based on the number of points $N$ and the number of time steps.
# Note that I choose to adapt the size of S or f to account for ghost cells before and after the solution domain to incorporate periodicity
# Next, I want to incorporate the temporal discretization to solve df/dt = -a df/dx, looping over this solution for each time step $n$
# The function should take as inputs a, L, N, $\Delta t$, the function f, and return the matrix f which gives the solution at each time step.
```

```python
def expEuler(dx,dt,a,fold,Spatial,N): #returns a vector of all update y-values
    dfdt = Spatial(fold,dx,a,N) #spatial takes in a vector of all f values and the time step size dx, and returns the value of dfdt at each spatial point x at the time tk+1.
    fn = np.add(fold,dt*dfdt)
    return fn

def RK3(dx,dt,a,fold,Spatial,N):
    K0 = Spatial(fold,dx,a,N)
    K1 = Spatial(fold+dt/2*K0,dx,a,N)
    K2 = Spatial(fold+2*dt*K1-dt*K0,dx,a,N)
    fn = fold + (K0+4*K1+K2)*dt/6
    return fn
```

```python
def gaussian(x):
    g = np.exp(-100*(x-0.5)**2)
    return g

def square(x):
    svec = np.zeros(np.shape(x))
    for xindex in range(np.size(x)):
        if ((0.25 < x[xindex]) and (x[xindex] <= 0.75)):
            svec[xindex] = 1
        else:
            svec[xindex] = 0
    return svec
```

```python
#spatial takes in a vector of all f values and the time step size dx, and returns the value of dfdt at each spatial point x at the time tk+1.

def BDA1(fold,dx,a,N):
    RHSvec = np.zeros(N+4)
    for xindex in np.arange(2,N+2,1):
        xindex = int(xindex)
        RHSvec[xindex] = -a*(fold[xindex]-fold[xindex-1])/dx
    RHSvec[0] = RHSvec[N]
    RHSvec[1] = RHSvec[N+1]
    RHSvec[N+2] = RHSvec[2]
    RHSvec[N+3] = RHSvec[3]
    return RHSvec

def Cent2(fold,dx,a,N):
    RHSvec = np.zeros(N+4)
    for xindex in np.arange(2,N+2,1):
        xindex = int(xindex)
        RHSvec[xindex] = -a*(fold[xindex+1]-fold[xindex-1])/(2*dx)
    RHSvec[0] = RHSvec[N]
    RHSvec[1] = RHSvec[N+1]
    RHSvec[N+2] = RHSvec[2]
    RHSvec[N+3] = RHSvec[3]
    return RHSvec

def Up3(fold,dx,a,N):
    RHSvec = np.zeros(N+4)
    for xindex in np.arange(2,N+2,1):
        xindex = int(xindex)
        RHSvec[xindex] = -a*(fold[xindex-2]-6*fold[xindex-1]+3*fold[xindex]+2*fold[xindex+1])/(6*dx)
    RHSvec[0] = RHSvec[N]
    RHSvec[1] = RHSvec[N+1]
    RHSvec[N+2] = RHSvec[2]
    RHSvec[N+3] = RHSvec[3]
    return RHSvec

def Cent4(fold,dx,a,N):
    RHSvec = np.zeros(N+4)
    for xindex in np.arange(2,N+2,1):
        xindex = int(xindex)
        RHSvec[xindex] = -a*(fold[xindex-2]-8*fold[xindex-1]+8*fold[xindex+1]-fold[xindex+2])/(12*dx)
    RHSvec[0] = RHSvec[N]
    RHSvec[1] = RHSvec[N+1]
    RHSvec[N+2] = RHSvec[2]
    RHSvec[N+3] = RHSvec[3]
    return RHSvec
```

```python
#We do points 0 thru N, where point x(N+1)=L=0
#f = F(:,0) should have N points
# This means F should have N+2 rows (with 2 ghost rows for 4th order central) 
# F should have numsteps = T/Delta T columns
# S should be NxN
# f0 is an argument of "solve"

def solveadv(a=1,L=1,N=50,f0=gaussian,CFL=0.5,Spatial=Cent2,TimeSolver=expEuler):
    
    #Spatial Controls
    dx = L/N # There are N points, comprising N-1 intervals, which cover L-dx of the domain. Thus, L-dx/N-1=dx, and L/N = dx.
    xvals = np.arange(0,L,dx)
    xvals.reshape(N,1)
    
    #Time Controls
    Tp = L/a
    runtime = 10*Tp
    dt = CFL*dx/a
    numsteps = int(runtime/dt)+1
    stepnums = np.floor(np.array([0,2*Tp,4*Tp,6*Tp,8*Tp,10*Tp])/dt)
    # stepnums = np.floor(np.array((0,100,200,300,400,500)))
    print(stepnums)
    
    # Initialize matrix F
    f0vals = f0(xvals)


    F = np.zeros((N+4,numsteps))
    F[2:N+2,0] = f0vals[0:N+1]
    F[0:2,0] = f0vals[N-2:N]
    F[N+2:N+4,0] = f0vals[0:2]
   
    
    
    # Loop over time to solve the problem
    for n in np.arange(1,numsteps,1):
        fold = F[:,n-1] #previous solution vector f
        # t = n*dt #current physical time
        # print('Time Step ' + str(n))
        fn = TimeSolver(dx,dt,a,fold,Spatial,N) #Note that TimeSolver will call the spatial discretization to evaluate spatial derivatives, and so it needs the fvals and the dx info.
        F[:,n] = fn
    
    #Get the solution at the appropriate times t=0,2*Tp,...
    Fplot = np.zeros((N+1,len(stepnums)+1))
    for stepindex in np.arange(0,len(stepnums),1):
        Fplot[0:N,0] = xvals[0:N]
        Fplot[N,0] = L
        Fplot[:,stepindex+1] = F[2:N+3,int(stepnums[stepindex])]
    return Fplot
```

```python
def plot_vals(Fplot,filename=[],title=''):
    x1 = Fplot[:,0]
    y1 = Fplot[:,1]
    y2 = Fplot[:,2]
    y3 = Fplot[:,3]
    y4 = Fplot[:,4]
    y5 = Fplot[:,5]
    y6 = Fplot[:,6]
    plt.figure(figsize=(10,5))
    plt.grid()
    axes_internal = plt.gca()
    axes_internal.set_xlabel('$x$')
    axes_internal.set_ylabel('$f$')
    if title: plt.title(title)
    plt.plot(x1,y1, label = '$t=0$')
    plt.plot(x1,y2, label = '$t=2T_p$')
    plt.plot(x1,y3, label = '$t=4T_p$')
    plt.plot(x1,y4, label = '$t=6T_p$')
    plt.plot(x1,y5, label = '$t=8T_p$')
    plt.plot(x1,y6, label = '$t=10T_p$')
    plt.legend()
    if len(filename) != 0:
        plt.savefig(filename+'.png')
    plt.show()
```

```python
Fplot = solveadv(a=1,L=1,N=50,f0=gaussian, CFL=0.5,Spatial=BDA1,TimeSolver=RK3)
plot_vals(Fplot,'GaussianBDA1')
Fplot = solveadv(a=1,L=1,N=50,f0=gaussian, CFL=0.5,Spatial=Cent2,TimeSolver=RK3)
plot_vals(Fplot,'GaussianCent2')
Fplot = solveadv(a=1,L=1,N=50,f0=gaussian, CFL=0.5,Spatial=Up3,TimeSolver=RK3)
plot_vals(Fplot,'GaussianUp3')
Fplot = solveadv(a=1,L=1,N=50,f0=gaussian, CFL=0.5,Spatial=Cent4,TimeSolver=RK3)
plot_vals(Fplot,'GaussianCent4')
```

```python
Fplot = solveadv(a=1,L=1,N=50,f0=square, CFL=0.5,Spatial=BDA1,TimeSolver=RK3)
plot_vals(Fplot,'SquareBDA1')
Fplot = solveadv(a=1,L=1,N=50,f0=square, CFL=0.5,Spatial=Cent2,TimeSolver=RK3)
plot_vals(Fplot,'SquareCent2')
Fplot = solveadv(a=1,L=1,N=50,f0=square, CFL=0.5,Spatial=Up3,TimeSolver=RK3)
plot_vals(Fplot,'SquareUp3')
Fplot = solveadv(a=1,L=1,N=50,f0=square, CFL=0.5,Spatial=Cent4,TimeSolver=RK3)
plot_vals(Fplot,'SquareCent4')
```

```python
def BDA1E(N):
    S = np.zeros((N,N))
    for i in np.arange(-1,N):
        for j in np.arange(-1,N):
            if i == j:
                S[i,j] = 1
            elif i == j + 1:
                S[i,j] = -1
    return S

def Cent2E(N):
    S = np.zeros((N,N))
    for i in np.arange(-1,N):
        for j in np.arange(-1,N):
            if i == j - 1:
                S[i,j] = 1
            elif i == j + 1:
                S[i,j] = -1
    return S

def Up3E(N):
    S = np.zeros((N,N))
    for i in np.arange(-2,N):
        for j in np.arange(-2,N):
            if i == j - 1:
                S[i,j] = 2
            elif i == j:
                S[i,j] = 3
            elif i == j + 1:
                S[i,j] = -6
            elif i == j + 2:
                S[i,j] = 1
    return S

def Cent4E(N):
    S = np.zeros((N,N))
    for i in np.arange(-2,N):
        for j in np.arange(-2,N):
            if i == j - 2:
                S[i,j] = -1
            elif i == j - 1:
                S[i,j] = 8
            elif i == j + 1:
                S[i,j] = -8
            elif i == j + 2:
                S[i,j] = 1
    return S
```
